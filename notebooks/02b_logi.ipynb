```r
library(tidyverse)
```

```r
data_dir <- file.path("..", "data")
```

```r
model <- readRDS(file.path(data_dir, "multivariate_model.rds"))
```

```r
coefs_mtx <- confint(model)
```

```r
coefs <- coefs_mtx %>%
as.data.frame(stringsAsFactors = FALSE) %>%
setNames(c("lo", "hi")) %>%
mutate(cov = rownames(.))

ests <- model %>%
summary %>%
pluck("coefficients") %>%
as.data.frame(stringsAsFactors = FALSE) %>%
setNames(c("est", "se", "zval", "pval")) %>%
mutate(cov = rownames(.))

df <- coefs %>%
full_join(ests, by = "cov")

df
```
